```python
def parse_raw_message(raw_message):
    lines = raw_message.split('\n')
    email = {}
    message = ''
    keys_to_extract = ['from', 'to']
    for line in lines:
        if ':' not in line:
            message += line.strip()
            email['body'] = message
        else:
            pairs = line.split(':')
            key = pairs[0].lower()
            val = pairs[1].strip()
            if key in keys_to_extract:
                email[key] = val
    return email
```

```python
#This function will ignore file that cannot be read as UTF8 encode
def readAllFile(path):
    result = []
    for root, dirs, files in os.walk(path):
        for file in files:
            try:
                with open(os.path.join(root, file), "r", encoding="ISO-8859-1") as auto:
                    result.append(auto.read())
            except:
                continue
    return result
```
